## Dados categóricos

## Dados Numéricos

```python
col_floats = list(df.iloc[:,2:-1].columns)
```

```python
## Dados nulos
df[col_floats] = df[col_floats].fillna(value=0)
```

```python
df[col_floats]
```

```python
df[col_floats].head()
```

```python
df[col_floats] = df[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0','').replace('.','').replace('%','').replace(',','.'))
```

# Criando uma função com uma estratégia para oportunidde de Mercado

```python
def oportunidade_media_setor_slug(df, setor_slug='shoppings', label_setor_slug='setor_slug'):
  media_setor_slug = df_aux.groupby('setor_slug').agg(['mean','std'])
  df_setor_slug = df[df[label_setor_slug].isin([setor_slug])]
  filter_ = \
          (df_setor_slug['Ativos'] >5 &\
           df_setor_slug['liquidezmediadiaria'] >0 &
           df_setor_slug['pvp'] <0 &
           df_setor_slug['media_yield_12m'] > media_setor_slug.loc[setor_slug, ('media_yield_12m', 'mean')])
```

```python
list(df['setor_slug'].unique())
```

```python
oportunidade = oportunidade_media_setor_slug(df_aux, setor_slug='imoveis-residenciais')
oportunidade.sort_values('media_yield_12m', ascending=False, inplace=True)
oportunidade
```
