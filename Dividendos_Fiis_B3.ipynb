```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl

class fii:
    def __init__(self, ticker):
        self.ticker = ticker
    def pegar_dividendos_fii(self):
        agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
        url = f"https://www.fundamentus.com.br/fii_proventos.php?papel={self.ticker}&tipo=2"
        webpage = requests.get(url, headers=agent)
        soup = BeautifulSoup(webpage.content, "html.parser")
        resultados = soup.find(id = "resultado")
        tabela_body = resultados.find("tbody")
        linhas = tabela_body.find_all("tr")
        data = []
        for linha in linhas:
            colunas = linha.find_all("td")
            colunas = [ele.text.strip() for ele in colunas]
            data.append([ele for ele in colunas if ele])
        data_output = pd.DataFrame(data, columns=["Data_com", "Rendto", "Data_Ultm_Rendto", "Valor"])
        data_output['Ticker'] = self.ticker
        # aqui pegaremos sempre o ultimo dividendo disponivel
        data_output = data_output.iloc[0]

        return data_output

def get_dividend(papel):
    return fii(papel).pegar_dividendos_fii()

# usando map para rodar a funcao para os tickers
resultados_map = pd.DataFrame(list(map(get_dividend, ['BBFI11B', 'BBPO11', 'BBRC11', 'BNFS11', 'CXAG11', 'RBOP11',
       'BTAL11', 'BTRA11', 'QAGR11', 'RZTR11', 'FAED11', 'FCFL11',
       'RBED11', 'BBGO11', 'CPTR11', 'CRAA11', 'DCRA11', 'EGAF11',
       'FGAA11', 'GCRA11', 'JGPX11', 'KNCA11', 'LSAG11', 'NCRA11',
       'RURA11', 'RZAG11', 'VCRA11', 'VGIA11', 'XPCA11', 'ALZM11',
       'BCFF11', 'BCIA11', 'BLMR11', 'BPFF11', 'CPFF11', 'CRFF11',
       'CXRI11', 'DVFF11', 'HFOF11', 'HGFF11', 'JSAF11', 'KFOF11',
       'MCHF11', 'MGFF11', 'NAVT11', 'OUFF11', 'RBFF11', 'RBRF11',
       'RFOF11', 'RVBI11', 'XPSF11', 'ALZR11', 'KNRI11', 'MAXR11',
       'MXRF11', 'RBRD11', 'RBRP11', 'RBVA11', 'RINV11', 'HCRI11',
       'HUSC11', 'NSLU11', 'HTMX11', 'XPHT11', 'MFII11', 'TGAR11',
       'FIIB11', 'XPIN11', 'ALMI11', 'BMLC11', 'BRCR11', 'BROF11',
       'CBOP11', 'CEOC11', 'CJCT11', 'CNES11', 'CTXT11', 'CXCE11B',
       'DRIT11B', 'EDFO11B', 'EDGA11', 'FAMB11B', 'FLMA11', 'FMOF11',
       'FPAB11', 'GTWR11', 'HGPO11', 'HGRE11', 'JSRE11', 'ONEF11',
       'PATC11', 'PRSV11', 'PVBI11', 'RCRB11', 'RECT11', 'RNGO11',
       'SPTW11', 'TEPP11', 'TRNT11', 'VINO11', 'XPCM11', 'XPPR11',
       'BLMG11', 'BRCO11', 'BTLG11', 'CXTL11', 'EURO11', 'GALG11',
       'GGRC11', 'HGLG11', 'HLOG11', 'HSLG11', 'LGCP11', 'LVBI11',
       'NEWL11', 'OULG11', 'PATL11', 'RBRL11', 'RELG11', 'RZAT11',
       'TRBL11', 'VILG11', 'XPLG11', 'AFHI11', 'BARI11', 'BCRI11',
       'BTCI11', 'CACR11', 'CPTS11', 'CVBI11', 'DEVA11', 'GCRI11',
       'HABT11', 'HCTR11', 'HGCR11', 'HSAF11', 'IRDM11', 'KCRE11',
       'KNCR11', 'KNHY11', 'KNIP11', 'KNSC11', 'MCCI11', 'NCHB11',
       'OUJP11', 'OURE11', 'PLCR11', 'PORD11', 'RBHY11', 'RBRR11',
       'RBRX11', 'RBRY11', 'RECR11', 'RRCI11', 'TORD11', 'URPR11',
       'VCJR11', 'VGIP11', 'VGIR11', 'VRTA11', 'WHGR11', 'XPCI11',
       'EVBI11', 'FIIP11B', 'HGRU11', 'TRXF11', 'JFLL11', 'ABCP11',
       'FIGS11', 'FLRP11', 'FVPQ11', 'GSFI11', 'HGBS11', 'HSML11',
       'MALL11', 'PQDP11', 'SHPH11', 'VISC11', 'VSHO11', 'WPLZ11',
       'XPML11'])))
# exportando FIIs
#resultados_map.to_excel("dividendos_fiis.xlsx")

#alterando posicionamento das colunas
resultados_map = resultados_map[['Data_com', 'Ticker', 'Rendto', 'Data_Ultm_Rendto', 'Valor']]
resultados_map
```
