# Criando uma função com uma estratégia para oportunidde de Mercado

```python
# FILTRO POR SETOR
setor = 'shoppings'

def oportunidade_media_setor_slug(df, setor_slug=None):
  label_setor_slug='setor_slug'
  media_setor_slug = df.groupby(label_setor_slug).agg(['mean','std'])

  df_setor_slug = df[df[label_setor_slug].isin([setor_slug])]

  # FILTRO POR INDICADORES
  filter_ = \
          (
           (df_setor_slug['ativos'] >= 0) &\
           (df_setor_slug['rentabilidade_mes'] >0) &\
           (df_setor_slug['pvp'] >0)
          )

  return df_setor_slug[filter_]

# RESULTADO
oportunidade = oportunidade_media_setor_slug(df_indicadores, setor)
oportunidade.sort_values('media_yield_12m', ascending=False, inplace=True)
oportunidade
```
