```python
# Análise de dados
import pandas as pd
import numpy as np
import vectorbt as vbt
import yfinance as yf
```

```python
# O abaixo retorna só Dividendos
tickers = ['BLMG11.SA','BRCO11.SA', 'BTLG11.SA','GALG11.SA', 'HGLG11.SA' ,'HLOG11.SA', 'LGCP11.SA', 'LVBI11.SA', 'NEWL11.SA', 'PATL11.SA', 'RBRL11.SA', 'SDIL11.SA', 'XPLG11.SA' ]
start_date = '2020-01-01'
# end_date =
#div = vbt.YFData.download(tickers, period='max', interval='1mo').get(['Dividends'])
div = vbt.YFData.download(tickers, start= start_date, interval='1d').get('Dividends')
# div.to_excel('FoFs_Dividends.xlsx')
# div.tail(10)
```

```python
div.index = pd.to_datetime(div.index.date)
# div
```

```python
div = div.dropna()
div.tail()
```

```python
close = yf.download(tickers, start= start_date, interval='1d')['Close']
close = close.dropna()
close.tail()
```

##2.Padronizando Dados

```python
div.index = pd.to_datetime(div.index.date)
close.index = pd.to_datetime(close.index.date)
```

```python
# classificando colunas
div = div.reindex(sorted(div.columns), axis= 1)
close = close.reindex(sorted(close.columns), axis= 1)
```

##3.Dividendo dos ultimos 12 Meses

```python
div_12m = div.rolling('365D').sum()
div_12m
```

##4.Dividend Yield 12m

```python
dy_12m = (div_12m/close)*100
dy_12m
```

##5.Plot

```python
dy_12m.vbt.plot(width= 1000, height= 600).show()
```
